```python
import pandas as pd
from google.colab import files
import os

# CARREGANDO MINHA PLANILHA
nome_arquivo = 'ProjetoETL.xlsx'

# Check if the file exists locally, if not, prompt for upload
if not os.path.exists(nome_arquivo):
  print(f"O arquivo '{nome_arquivo}' não foi encontrado. Por favor, faça o upload.")
  uploaded = files.upload()
  if nome_arquivo not in uploaded:
    print(f"O arquivo '{nome_arquivo}' não foi carregado. Certifique-se de fazer o upload do arquivo correto com o nome '{nome_arquivo}'.")
    # Exit or raise an error if the file still doesn't exist
    raise FileNotFoundError(f"Arquivo '{nome_arquivo}' não encontrado após o upload.")


df = pd.read_excel(nome_arquivo)

# REMOVENDO ESPACOS EM BRANCO
df = df.applymap(lambda x: x.strip() if isinstance(x, str) else x)

# EXPORTAR A PLANILHA POS ORGANIZAÇÃO DOS DADOS
df.to_excel('planilha_limpa.xlsx', index=False)

print("Processamento concluído!")
```

```python
#APLICAR UM DESCONTO PARA VALORES MAIORES QUE 500

# Criando a coluna 'Desconto' inicializada com 0
df["Desconto"] = 0

# SE o valor for maior que 500, aplicamos o desconto de 10%

df.loc[df["VALOR "] > 500, "Desconto"] = df["VALOR "] * 0.10

# Calculando o valor com desconto
df["Valor_Final"] = df["VALOR "] - df["Desconto"]

# Salvar o resultado
df.to_excel("planilha_com_desconto.xlsx", index=False)

print("Desconto aplicado com sucesso!")
```
